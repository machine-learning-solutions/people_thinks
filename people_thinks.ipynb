# 2) Preprocessing





```python
import re
def Arabic_Normalizing(text):
    text = re.sub("انا","", text)    
    text = re.sub("من","", text)
    text = re.sub("الى","", text)
    text = re.sub("عن","", text)
    text = re.sub("على","", text)
    text = re.sub("في","", text)
    text = re.sub("لكن","", text)
    text = re.sub("لعل","", text)
    text = re.sub("حتى","", text)
    text = re.sub("كان","", text)
    text = re.sub("لا","", text)
    text = re.sub("ابو","", text)
    text = re.sub("اللهم","", text)
    text = re.sub("صل","", text)
    text = re.sub("وسلم","", text)
    text = re.sub("على","", text)
    text = re.sub("سيدنا","", text)
    text = re.sub("نبينا","", text)
    text = re.sub("وحبيبنا","", text)
    text = re.sub("محمد","", text)
    text = re.sub("ابوك","", text)
    text = re.sub("بل","", text)
    text = re.sub("هل","", text)
    text = re.sub("عل","", text)
    text = re.sub("انت","", text)
    text = re.sub("[إأآا]", "ا", text)
    text = re.sub("ى", "ي", text)
    text = re.sub("ؤ", "ء", text)
    text = re.sub("وا", "و", text)
    text = re.sub("ئ", "ء", text)
    text = re.sub("ة", "ه", text)
    text = re.sub("گ", "ك", text)
    return text
arabic_diacritics = re.compile("""
                             ّ    | # Tashdid
                             َ    | # Fatha
                             ً    | # Tanwin Fath
                             ُ    | # Damma
                             ٌ    | # Tanwin Damm
                             ِ    | # Kasra
                             ٍ    | # Tanwin Kasr
                             ْ    | # Sukun
                             ـ     # Tatwil/Kashida
                         """, re.VERBOSE)
def remove_diacritics(text):
    text = re.sub(arabic_diacritics, '', text)
    return text
def filter_stop_words(train_sentences, stop_words):
    for i, sentence in enumerate(train_sentences):
        new_sent = [word for word in sentence.split() if word not in mystopwords_list]
        train_sentences[i] = ' '.join(new_sent)
    return train_sentences   
new1=[]
new2=[]
new3=[]
new4=[]
new5=[]
for a in range(len(df)):
    ## Drop non words
    pattern1 = r'[^\w]'
    ## Drop non Arabic words
    pattern2 = r'[a-zA-Z?]'
    ## Drop Numbers
    pattern3 = r'[0-9٠-٩?]'

    new1.append(re.sub(pattern1, ' ', str(df[0][a])))
    new2.append(re.sub(pattern2, ' ', str(df[0][a])))
    new3.append(re.sub(pattern3, ' ', str(df[0][a])))
    new4.append(Arabic_Normalizing(str(df[0][a])))
    new5.append(remove_diacritics(str(df[0][a])))

a_series = pd.Series(new1)
b_series = pd.Series(new2)
c_series = pd.Series(new3)
d_series = pd.Series(new4)
e_series = pd.Series(new5)
df = pd.concat([a_series, b_series,c_series,d_series,e_series], axis=0,ignore_index=True)

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_list = stopwords.words('arabic')
mystopwords_list = stopwords_list

mynewlist = ["هيك" , "هيكا" , "إزا" , "اه" , "ااه"  , "اها",  "اهاا" , "هاي", "هون","هونا", "هدول", "هظول","هضول" ,"هدول" ,"مش"  , "ولا", "اشي" ,
 "إشي" , "اياه" , "الكم" , "إلكو" , "الكو" , "شو" , "إيش" , "ايشي" , "يللي" , "فش" , "فيش" , "ع" , "اي",  "أي",  "ليش",  "يلا" ,
 "يللا" , "النا" , "إلنا" , "إلكم" , "شو" , "احنا" , "إحنا" , "ول" , "إنتو", "إنتم" , "إنت" , "إنتا" ,"إنتي" , "أوف",  "أف" , "هاظ" ,
 "هاد" , "هاظا" , "هادا"  "هاذا" , "بس",  "شسمو"  , "هضلاك"  , "هظلاك" , "أوه" ,"ه","شهاد" ,"تا" , "عشان" , "عشنه" ,"عشنو"  , "عشنكم" , "كتير" , "حدا" , "حد"  ] 

unicode_range = ('0025','00FF')
characters = []
for unicode_character in range(int(unicode_range[0] , 16), int(unicode_range[1] , 16)):
    characters.append(chr(unicode_character))

for i in mynewlist :
      mystopwords_list.append(i)


df = filter_stop_words(df, mystopwords_list)
df.drop_duplicates(keep='first',inplace=True)
```

# 3) Creating A model



## A) Split Text

```python
listOfComment = []
for row in df:
            mylist = row.split()
            listOfComment.append(mylist)
```

## B) Creating Vocabulary

```python
vocab = {}
count = 0
for doc in df:
    for word in doc.split():
        if word not in vocab:
            count = count +1
            vocab[word] = count
```

## C) Word Embedding

```python
from gensim.models import Word2Vec

word2vec = Word2Vec(listOfComment)
vocab_len = len(word2vec.wv)
vocab_len
```
